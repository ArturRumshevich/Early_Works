```python
#для удобства занесем информацию по уровню образования в словарь:
dict_education = {'1' : 'начальное общее образование',
                  '2' : 'основное общее образование',
                  '3' : 'среднее общее образование',
                  '4' : 'среднее профессиональное образование',
                  '6' : 'высшее образование',
                  '9' : 'другое',
                  '0' : 'нет данных'
                 }
```

```python
#отобразим информацию из словаря в данных:
df['education'] = df['education'].map(dict_education)
df.head(10)
```

```python
#Соотнесем числовые значения с описанием пола и применим словарь к данным:
dict_gender = {'1' : 'мужской',
               '0' : 'женский',
               'None' : 'нет данных'
              }
df['gender'] = df['gender'].map(dict_gender) 
df.head()
```

```python
#Посмотрим на людей, ранее официально не работавших по уровню образования:
did_not_work = df[df['hc_didnotwork'] == True].pivot_table(index='education',
                        aggfunc='sum',
                        values='count',
                        ).reset_index().sort_values(by='count', ascending=False)
did_not_work
```

*По таблице можно сделать вывод о том, что больше всего работу ищут люди со средним общим образованием, то есть люди, окончившие 11 классов. Это может быть связано с тем, что выпускники уже хотят начать зарабатывать, но у них для этого не хвтает профессионального образования и опыта работы.*

```python
#Теперь сделаем то же самое только для людей, которые трудоустроились:
found_education = df[df['find_new_job'] == True].pivot_table(index='education',
                        aggfunc='sum',
                        values='count',
                        ).reset_index().sort_values(by='count', ascending=False)
found_education
```

```python
#Сейчас выясним общее количество людей по уровню образования:
count_education = df.pivot_table(index='education', aggfunc='sum', values='count').reset_index().sort_values(by='count', ascending=False)
count_education
```

```python
#Сейчас можно поделить одни значения на другие, чтобы получить проценты:
found_education['count'] = round(found_education['count'] / count_education['count'] * 100, 2)
found_education
```

```python
#Визуализируем в виде столбчатой диаграммы соотношение трудоустройства по каждому уровню образования:
sns.barplot(data=found_education, x='education', y='count').set(title='Процент трудоустроившихся по уровню образования',
                                                                 xlabel='Уровень образования', 
                                                                 ylabel='Процент трудоустроившихся'
                                                               )
plt.rcParams['figure.figsize'] = (20, 20)
plt.show()
```

*Вывод по 5 заданию: больше всего людей трудоустроились с высшим образованием и затем со средним профессиональным.*

```python
#Перейдем к исследованию безработицы в 2020 году.
data_1 = df.query('month_application >= "2020-01-01" and month_application < "2020-12-01"')
data_1.head()
```

```python
#Сейчас посмотрим на динамику трудоустроившихся по месяцам:
report8 = data_1[data_1['find_new_job'] == True].pivot_table(index='month',
                        aggfunc='sum',
                        values='count',
                        ).reset_index()
dict_months = {1: 'Январь', 2: 'Февраль', 3: 'Март', 4: 'Апрель', 5: 'Май', 6: 'Июнь', 7: 'Июль', 8: 'Август', 9: 'Сентябрь', 10: 'Октябрь', 11: 'Ноябрь'}
report8['month'] = report8['month'].map(dict_months)

report9 = data_1.pivot_table(index='month', aggfunc='sum', values='count').reset_index()
report8['not_found'] = report9['count'] - report8['count']

report8.plot(kind='bar', x='month', y=['count', 'not_found'], title='Динамика трудоустроившихся и не трудоустроившихся за 2020 год', xlabel='Месяц', ylabel='Количество трудоустроившихся и не трудоустроившихся')
plt.show()
```

*Вывод по пункту 6.1: Детализируя информацию о безработных и трудоустроившихся за 2020 год, можно сделать вывод о том, что, начиная с апреля 2020 года, процент нетрудоустроенных резко возрастает, затем этот процент снижается, но тем не менее разрыв между трудоустроенными и нетрудоустроенными остается колоссальным.
